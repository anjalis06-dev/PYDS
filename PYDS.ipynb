Numpy : Numerical Python

Important attributes of Numpy

1. NumPy array is called as **ndarray**
2. **ndarray.ndim**: the no of axes(dimensions) of the arrays
3. **ndarray.shape** : Tuple of integers that gives size of the array in each dimension
for an m n matix, shape is (n,m). the length of the shape of tuple is number of axes, ndim.
4.** ndarray.size**: the total number of elements of the array is equal to the product of the elements of the shape.
5. **ndarray.dtype**: An Object that descibes the type of elements in the array
For exm: numpy.int32, numpy.int16, numpy.float64

6. **ndarray.itemsize**: size in bytes of each element of the array
For exm : elements are float64 then
itemsize= 64/8 = 8
for complex 32, itemsize = 32/8 = 4
 it is denoted by **ndarray.dtype.itemsize**

 7. **ndarray.data** : the buffer contains the actual elements of array. However, for accessing the elements of the array , use indexing and not ndarray.data

Conditional Handling in Numpy Python
1. **np.where** =  function is a vectorized tool used for conditional operations on arrays.

*  Finding Indices (One Argument)When used with only a condition, np.where(condition) returns a tuple of arrays containing the indices where the condition is True.
* Conditional Replacement (Three Arguments)When used with three arguments, np.where(condition, x, y), it acts like a vectorized if-else statement.


2. **np.select** =  is to perform complex conditional mapping on an array where multiple different outcomes are required based on multiple conditions.


```python
x = np.array([i for i in range(10)])
```

```python
x
```

```python
np.where(x%2==0,'Even','Odd')
```

```python
#condition list
condlist = [x<5,x>5]
#choice list
choicelist=[x**2,x**3]
```

```python
np.select(condlist,choicelist,default=x)
```

**Linear Algebra**:

```python
def dt_product(x,y):
  return sum(i*j for i,j in zip(x,y,strict=True))
```

```python
dt_product([1,5,7],[8,9,6])
```

```python
#norm of vector == length of vector
import math
def norm_of_vector(v):
  dot_product= sum(i*i for i in v)
  return math.sqrt(dot_product)
norm_of_vector([8,9,6])
```
